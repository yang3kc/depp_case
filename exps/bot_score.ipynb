```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'axes.axisbelow': True})
```

# Load data

```python
blt_score_df = pd.read_csv(
    "/home/yangkc/depp_case/data/intermediate_data/bot_related/user_blt_scores.csv.gz",
    dtype={"user_id": "O"}
)
```

```python
tweet_df = pd.read_parquet("../data/intermediate_data/tweet_info/tweet_info_all.parquet")
```

```python
tweet_score_df = tweet_df.merge(blt_score_df[['user_id', 'bot_score_lite']], on='user_id')
```

# Show results

```python
plt.hist(blt_score_df.bot_score_lite, bins=20)
plt.xlabel("Bot score")
plt.ylabel("Frequency")
plt.gca().axvline(0.5, color='red')
plt.tight_layout()
```

```python
plt.hist(tweet_score_df.bot_score_lite, bins=20)
plt.xlabel("Bot score")
plt.ylabel("Frequency")
plt.gca().axvline(0.5, color='red')
plt.tight_layout()
```
