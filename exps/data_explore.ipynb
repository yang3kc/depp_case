# For bot detection

```python
one_tweet_per_user_path = "/home/yangkc/depp_case/data/intermediate_data/bot_related/user_tweet.ndjson"
```

```python
known_users = set()
with open(one_tweet_per_user_path, "w") as f:
    counter = 0
    for tweet in tweets:
        if tweet['user']['id_str'] not in known_users:
            temp_tweet = {
                'user': tweet['user'],
                'created_at': tweet['created_at']
            }
            str_to_write = json.dumps(temp_tweet)
            f.write(f"{str_to_write}\n")
            known_users.add(tweet['user']['id_str'])
        if counter % 1000 == 0:
            print(f"{len(known_users)} / {counter}")
        counter += 1
```
