# Split the list

```python
def chunk_df(df, chunk_size):
    """Splits the list 'lst' into chunks of size 'chunk_size'."""
    return [df.iloc[i : i + chunk_size] for i in range(0, len(df), chunk_size)]
```

```python
chunks = chunk_df(url_df, 10000)
```

```python
len(chunks)
```

```python
for index, chunk in enumerate(chunks):
    chunk.to_csv(
        f"/home/yangkc/depp_case/data/intermediate_data/profile_images/urls/url_chunk_{index:0>2}.csv.gz", index=None
    )
```
